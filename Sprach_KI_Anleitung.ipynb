## Ziel des Projektes

Herzlich willkommen im Jupyter Notebook aus dem Projekt „Sprach-KI zum Anfassen“. Ziel des Projekts ist es, den Schülern ein tiefgehendes Verständnis und praktische Kenntnisse im Umgang mit lokalen Large Language Models (LLMs) zu vermitteln. Die Schüler sollen die grundlegenden Konzepte und Funktionsweisen von LLMs verstehen lernen, einschließlich der verschiedenen Arten von Modellen und deren Anwendungen. Durch interaktive Anleitungen und Übungen werden sie in die Lage versetzt, LLMs selbstständig zu installieren und zu konfigurieren. Dies schließt auch das Einrichten von notwendiger Software und die Nutzung von Schnittstellen wie Jupyter Notebooks und Docker ein. Zudem werden die Schüler erfahren, wie sie LLMs für spezifische Aufgaben wie Textgenerierung und Sprachverarbeitung einsetzen können. Ein zentrales Lernziel ist die Vermittlung von Anpassungsmethoden wie Userprompts, Systemprompts und Retrieval Augmented Generation (RAG), um die Schüler zu befähigen, LLMs effektiv in ihren Projekten und bei realen Problemstellungen einzusetzen. Durch diese Ziele sollen die Schüler nicht nur technische Fähigkeiten erwerben, sondern auch ein umfassendes Verständnis dafür entwickeln, wie diese revolutionäre Technologie in der Praxis genutzt werden kann.

- Im folgenden seht ihr das Installationsskript für Ubuntu.

- Ihr könnt dasselbe Skript auch für Debian benutzen, müsst aber an jeder Stelle “Ubuntu” mit “Debian” ersetzen.

<div style="display: flex; align-items: stretch; width: fit-content; margin: 20px; border-radius: 15px; overflow: hidden; border: 2px solid #cca000; background: rgba(204, 160, 0, .1) ">

```python
##Ubuntu

sudo apt-get update
sudo apt-get install ca-certificates curl
sudo install -m 0755 -d /etc/apt/keyrings
sudo curl -fsSL https://download.docker.com/linux/ubuntu/gpg -o /etc/apt/keyrings/docker.asc
sudo chmod a+r /etc/apt/keyrings/docker.asc

# Als nächstes wird das Installationsverzeichnis von Docker zu apt hinzugefügt:
echo \
  "deb [arch=$(dpkg --print-architecture) signed-by=/etc/apt/keyrings/docker.asc] https://download.docker.com/linux/ubuntu \
  $(. /etc/os-release && echo "$VERSION_CODENAME") stable" | \
  sudo tee /etc/apt/sources.list.d/docker.list > /dev/null
sudo apt-get update
sudo apt-get install docker-ce docker-ce-cli containerd.io docker-buildx-plugin docker-compose-plugin
```

- Als letztes muss der eigene Benutzer noch zur Gruppe Docker hinzugefügt werden.

- Um diese Änderung anzuwenden muss das System danach neugestartet werden.

```python
sudo usermod -aG docker <Eigener Benutzer>
sudo reboot
```

<div style="margin-left: 20px;">

Da wir nun Ollama erfolgreich installiert haben, können wir damit beginnen ein Modell herunterzuladen.

Für diesen Schritt verwenden wir die folgende Codezelle. Lass dich bitte nicht vom Code einschüchtern – es genügt, wenn du einfach die Zelle ausführst:

</div>

```python
from tqdm import tqdm
from ollama import pull


current_digest, bars = '', {}
for progress in pull(modell, stream=True):
  digest = progress.get('digest', '')
  if digest != current_digest and current_digest in bars:
    bars[current_digest].close()

  if not digest:
    print(progress.get('status'))
    continue

  if digest not in bars and (total := progress.get('total')):
    bars[digest] = tqdm(total=total, desc=f'pulling {digest[7:19]}', unit='B', unit_scale=True)

  if completed := progress.get('completed'):
    bars[digest].update(completed - bars[digest].n)

  current_digest = digest

print('Du kannst nun mit dem Tutorial fortfahren!')
```

### Mögliche Komplikationen

Auftauchende Fehlermeldung: 

<p style= "border:2px solid red;border-radius:2px;text-align:center; width: fit-content; padding: 7px; border-radius: 10px; background: rgba(255, 0, 0, .1)">b'docker: Error response from daemon: Ports are not available: exposing port TCP 0.0.0.0:3000 -> 0.0.0.0:0: listen tcp 0.0.0.0:3000: bind: Only one usage of each socket address (protocol/network address/port) is normally permitted.\n' </p>

Falls der gewählte Port belegt ist, muss ein anderer Port für den Container gewählt werden. Dafür kann der untenstehende Code ausgeführt werden. Es erscheint ein Eingabefenster, in das eine Portnummer eingetippt werden muss. Dann drückt "Enter":

```python
port = input("Bitte eine Zahl ab 3000 eingeben:") # Portnummer speichern
arg = shlex.split("docker rm open-webui") # Kommando: alter Open Web-UI Container wird gelöscht
shell_out= subprocess.run(arg, shell=True, capture_output=True) # Kommando ausführen: alter Open Web-UI Container wird gelöscht
print(shell_out.stderr) 

# Neuen Container mit neuen Port installieren
arg = shlex.split(f"docker run -d -p {port}:8080 --add-host=host.docker.internal:host-gateway -v open-webui:/app/backend/data --name open-webui --restart always ghcr.io/open-webui/open-webui:main")
shell_out = subprocess.run(arg, shell=True, capture_output=True)
print(shell_out.stderr)
```

Falls derselbe Fehler nochmal auftritt, heißt das, dass man wieder einen belegten Port gewählt hat. Um einen anderen Port zu wählen, kann die obere Zelle erneut ausgeführt werden.

Zuerst müssen wir alle notwendigen Pakete importieren

```python
# Importieren der Pakete
import nltk
from nltk.tokenize import word_tokenize
from gensim.models import Word2Vec
import matplotlib.pyplot as plt
import numpy as np
import torch
import torch.nn.functional as F
import math
from transformers import AutoModelForCausalLM, AutoTokenizer, set_seed
from deep_translator import GoogleTranslator
```

##### Tokenisieren

Der Input wird kleinere Einheiten zerlegt, die dann Tokens heißen. Die Tokens in diesem Beispiel wären: "Was", "ist", "der", "Satz", "des", "Pythagoras", "?".

**Input:** Satz

**Output:** Tokens als Liste


Codebeispiel: 

Wir definieren einen Input in der Variable "input_text" und übergeben ihn der Funktion "tokenisieren".

```python
# Eingabesatz
input_text = "Erkläre mir den Satz des Pythagoras, als würdest du es einem Kind erzählen."
```

```python
# Eine Funktion wird zuerst mit "def" definiert
def tokenisieren(input_text: str):
    # Laden notwendiger Ressourcen von NLTK
    nltk.download('punkt')

    # Tokenisierung in Wörter
    tokens = word_tokenize(input_text)

    # Ausgabe der Tokens
    print(tokens)
    return tokens

# Funktion wird ausgeführt, indem der Prompt (input_text) zum Tokenisieren übergeben wird
tokens = tokenisieren(input_text)

# source: https://www.nltk.org/api/nltk.tokenize.html 
```

##### Embedding

Um sinnvolle Ergebnisse zu generieren muss das Modell rechnen. Da man mit Zeichen schlecht rechnen kann, muss eine Umwandlung in Zahlen erfolgen.

Ein Embedding ist die Repräsentation eines Tokens (Buchstabe, Wort, Satzzeichen) als ein Zahlenvektor.

Wörter mit ähnlicher Bedeutung haben auch ähnliche Embeddings. Zum Beispiel sind die Wörter "Formel" und "Satz des Pythagoras" als Embeddings sehr ähnlich.


**Input:** Tokens als Liste

**Output:** Vektoren von den einzelnen Tokens bzw. Embeddings

Quelle: [YouTube-Video von StatQuest erklärt ab 1:30 wie Embeddings erstellt werden](https://youtu.be/zxQyTK8quyY?si=dugdu-rJlLCvr2kX)

Codebeispiel:

Wir verwenden die Tokens aus Schritt 1 und wandeln sie in ein Vektor um. 

```python
# Nochmal zur Erinnerung: Eine Funktion wird zuerst mit "def" definiert
def embedden(tokens:list):
    # Tokens matrisieren und Länge speichern für das Training des Embedding Modells
    token_list_len = len(tokens)
    tokens_matrix = [tokens]
    
    # Training des Word2Vec-Modells (Embedding Modell)
    model = Word2Vec(tokens_matrix, vector_size=token_list_len, window=5, min_count=1, workers=4)
    tokens_list = tokens_matrix[0]
    embeddings = []
    for word in tokens_list:
        vektor = model.wv[word]
        # Ausgabe des Vektors
        print(f"Vektor für das Wort '{word}': {vektor}")
        embeddings.append(model.wv[word])
    return embeddings

embeddings = embedden(tokens)

# source: https://radimrehurek.com/gensim/auto_examples/tutorials/run_word2vec.html
```

#### Positionelles Kodieren

Die Wortposition ist sehr wichtig, um einen sinnvollen Output zu erhalten. Betrachten wir folgende Sätze mit den gleichen Worten: 1. Lisa isst eine Pizza., 2. Eine Pizza isst Lisa. Beide Sätze haben aufgrund der Position ihrer Wörter völlig unterschiedliche Bedeutungen. Die Reaktion zum 1. Satz wäre, z. B. "Lecker!", wobei  die Reaktion zum 2. Satz z.B. "Was?!" wäre. 

Vereinfacht gesagt: Zu dem Index der Wörter (z. B. 0 = Was, 1 = ist, 2 = der, 3= Satz, 4= des, 5 = Pyhtagoras, 6 = ?) wird jetzt eine Positions-Zahl (Nennen wir ihn: p-Wert) addiert, die die Position des Wortes im Satz repräsentiert. 

Woher kommt jetzt dieser p-Wert? Zum Kodieren stehen uns eine Reihe von verschiedenen trigonometrischen Funktionen zur Verfügung (z. B. sin(4*x) und 3*cos(5*x)). Jeder Index ist als x-Wert auf diesen Funktionsgraphen zu verstehen. Der zum x-Wert korrespondierende y-Wert ist am Ende unser p-Wert. 



Bild-Quelle: https://uvadlc-notebooks.readthedocs.io/en/latest/tutorial_notebooks/tutorial6/Transformers_and_MHAttention.html 

Nach der Kodierung addiert man die Positionscodes mit den Embeddings und erhält ein Embedding, die nicht nur eine Wortrepräsentation als Zahl ist sondern auch die Position des Wortes repräsentiert.

Warum verwendet man trigonometrische Funktionen? 

Durch das periodische Verhalten der Funktion kann man bestimmte Positionen, wie Satzanfang oder -ende, generalisieren. Gleichzeitig kann man auch vielfältige Positionen durch die auf- und absteigenden Kurven erhalten.

**Input:** Embeddings

**Output:** Embeddings mit Positions-Kodierung 

Quelle: [YouTube-Video von StatQuest, dass ab 7:31 erklärt, wie Positionelles Kodieren funktioniert](https://youtu.be/zxQyTK8quyY?t=451)

Codebeispiel:

```python
def positionellesKodieren(embeddings):
    seq_len = len(embeddings)
    d = seq_len # Dimension der Embeddings
    n = 100000 # Feinheitsgrad der Kodierung. Je höher n ist, desto feiner
    P = np.zeros((seq_len, d)) # Leere Matrix in der Dimension der Matrix mit den Embeddings
    dimension_indices = np.arange(int(d/2)) # Indizes der Matrix, die konkret für die Berechnung unten notwendig sind
    # ab hier beginnt die iterative Berechung
    for k in range(seq_len):
        for i in dimension_indices:
            nenner = np.power(n, 2*i/d)
            P[k, 2*i] = np.sin(k/nenner) # Sinusfunktion für die Berechung des  p-Wertes
            P[k, 2*i+1] = np.cos(k/nenner) # Cosinusfunktion für die Berechnung des p-Wertes
    # Ausgabe der Positionscodes
    for i, w in enumerate(P):
        print(f"Positionelle Codes für {i+1}.tes Token: {P[i]}")
    return P

P = positionellesKodieren(embeddings)


def addiereMatrizen(m1, m2):
    new_m= np.add(m1, m2)
    for i,w in enumerate(new_m):
        print(f"\nNeuer Vektor für {i}.tes Token: {new_m[i]}")
    return new_m

pos_cod_matrix = addiereMatrizen(P,embeddings)
```

#### Attention

Während wir vorhin die Positionen der Wörter kodiert haben müssen wir nun die Bedeutung der Wörter in ihrem jeweiligen Kontext kodieren.
Manche Wörter können nur aus dem Kontext heraus verstanden werden. Das ist zum Beispiel bei Pronomen im Satz der Fall, die an Stelle eines Nomen stehen, das zuvor erwähnt wird. In unserer Eingabe "Erkläre mir den Satz des Pythagoras, als würdest du es einem Kind erzählen." ist das Pronomen "es" ein Fürwort für "Satz des Pythagoras". Diesen Zusammenhang erlernt ein Modell durch Attention.



[Bildquelle](https://jalammar.github.io/illustrated-transformer/)

Auf dem Bild sind die Ergebnisse eines Attention-Prozesses von einer bestimmten Eingabe dargestellt. Man sieht an den stark gefärbten orangenen Linien, wie das Pronomen "it" korrekt mit "the animal" assoziiert wird.  

Warum heißt Attention "Attention"?

Beim Attention werden nicht alle Wörter im Satz miteinander verglichen, sondern nur die wichtigsten Wörter. Zum Beispiel sind in unserer Frage sind die wichtigsten Wörter "Pythagoras", "es", "Kind" und "erzähle". Es reicht aus diese Wörter zu verstehen, um einen sinnvollen Output zu generieren.

**Input:** Matrix mit Embedding und Positions-Codierung vom Input und vom Output  

**Output:** Matrix des Outputs mit ihren Wahrscheinlichkeiten

Quelle: [Artikel: Attention is all you need](https://arxiv.org/abs/1706.03762)


Codebeispiel:

Hier werden nur die Attention-Werte vom Input berechnet (Self-Attention). In der Praxis werden noch die Attention-Werte des gesamten möglichen Outputs (also Trainingsdatensatz) berechnet. Anschließend werden dann noch die Attentionen des Input mit dem Output berechnet (Encoder-Decoder Attention). Dann haben wir ein Reihe von möglichen Output-Werten mit einer Wahrscheinlichkeit.

```python
def attention(pos_cod_matrix, mask=None):
    # Matrix zum Tensor umwandeln
    tensor = torch.from_numpy(pos_cod_matrix)
    # Aus diesen Parametern wird die Aufmerksamkeit berechnet
    query = tensor # Davon wollen wir die Aufmerksamkeiten erfragen 
    key = tensor # Damit wollen wir query vergleichen um Aufmerksamkeitswahrscheinlichkeiten zu generieren
    value = tensor # Damit werden am Ende die Aufmerksamkeitsgewichte multipliziert, 
                    # so dass man eine Ausgabe der Aufmerksamkeiten von allen Embeddings erhält
    
    # Generierung von wahrscheinlichen Aufmerksamkeitsgewichten
    d_k = query.size(-1)
    scores = torch.matmul(query, key.transpose(-2, 1)) / math.sqrt(d_k)
    if mask is not None:
        scores = scores.masked_fill(mask == 0, float('-inf'))
    # Softmax-Funktion klassifiziert und gibt finale Aufmerksamkeitsgewichte aus
    attention_weights = F.softmax(scores, dim=-1)
    # Matrizenpultiplikation von value und Aufmerksamkeitsgewichten
    output = torch.matmul(attention_weights, value)
    print("Aufmerksamkeitsgewichtsmatrix für die Embeddings. Wird klassisch so für Berechnungen benutzt. \nPro Zeile stehen die Aufmerksamkeiten für ein Embedding:\n", output)
    return output


input_att = attention(pos_cod_matrix)
```

#### Dekodieren

Der Output wird in diesem Abschnitt klassifiziert und in ein Text dekodiert bzw. übersetzt. Nach dem Attention hat man zunächst eine Menge an möglichen Outputs. Sie alle haben unterschiedliche Wahrscheinlichkeiten.
Das heißt, das zum Beispiel zu einem bisher generierten Satz "Der Hund.." mehrere mögliche Fortsetzungen möglich sind: "bellt" (Wahrscheinlichkeit 0,9), "sitzt" (Wahrscheinlichkeit 0,5) oder "spielt" (Wahrscheinlichkeit 0,7). Hierbei wird die Auswahl auf "bellt" treffen, weil es die höchste Wahrscheinlichkeit hat. Aber wir dürfen nicht vergessen, dass all diese Vorgänge mit Embeddings und nicht mit echten Wörtern durchgeführt werden. 




Es gibt einen bekannten Algorithmus zur Dekodierung in einen Text: 

    Top-K Sampling: Wählt aus einer k-fachen Menge wahrscheinlicher Outputs einen zufälligen Output. Da man meistens viele ähnlich wahrscheinliche Outputs hat beschränkt man sich auf k-Outputs und trifft von denen eine zufällige Auswahl. 

Hier gibt es einen interessanten Einstellungsparameter: 

    Temperatur: Je höher desto ungewöhnlicher sind die Outputs. Hat einen Wertebereich zwischen 0 und 1. Bei 0.3 hat man noch einen moderaten Output zu erwarten.

Probiert im unteren Codebeispiel verschiedene Temperaturen aus!


Nachdem ein geeigneter Text in Vektorrepräsentation zusammengestellt wurde muss alles noch in Wörter umgewandelt werden.

**Input:** Matrix des Outputs mit Wahrscheinlichkeiten

**Output:** Text

Quelle: [HuggingFace stellt die 3 Methoden nochmal anschaulich da](https://huggingface.co/blog/how-to-generate)

Codebeispiel:

Für die Dekodierung brauchen wir ein trainiertes Sprachmodell. Zur Demonstration der Dekodierungsalgorithmen verwende ich das kleine und leistungsschwache Modell GPT-2. Das ist schnell und ressourcenarm. 

```python
temperature = 0.8 # Hier könnt ihr eine Temperatur zwischen 0 - 1 wählen
```

```python
def decode():

    # Ausgabe-Modell 'gpt2' und sein Tokenizer werden geladen. Der Input wird tokenisiert mit dem GPT2-Tokenizer
    torch_device = "cuda" if torch.cuda.is_available() else "cpu" # Wahl det CPU als Prozessor, wenn es keine NVIDIA-GPU gibt 
    tokenizer = AutoTokenizer.from_pretrained("gpt2")
    model = AutoModelForCausalLM.from_pretrained("gpt2").to(torch_device)
    model_inputs = tokenizer(input_text, return_tensors='pt').to(torch_device)

    # Sampling-Methode
    set_seed(48) # Je höher der Seed desto schwerer ist das Verhalten des Modells reproduzierbar
    sample_output = model.generate(
        **model_inputs,
        max_new_tokens=40,
        do_sample=True,
        top_k=30,
        temperature = temperature,
    )
    out_sampling = tokenizer.decode(sample_output[0], skip_special_tokens=True)

    # Output kann Englisch sein, deshalb wird ein Übersetzer geladen
    translator = GoogleTranslator(source='en', target='de')

    # Ausgabe
    print("\nOutput mit Sampling:\n" + 100 * '-')
    print(translator.translate(out_sampling))
    return out_sampling

sampling_text = decode()
```

Wenn wir uns den Output der drei Methoden genau anschauen erkennen wir, dass mit Sampling die besten Ergebnisse gewonnen werden können. Denn es wird ein völlig neuer Text ohne Wiederholungen produziert. 

Trotzdem ist der Output nicht sehr sinnvoll. Das liegt daran, dass wir hier ein sehr leichtgewichtiges Modell "GPT-2" verwenden, das auf weniger Parameter (Trainingsdaten) trainiert wurde. Zum Vergleich: GPT-2 hat 137 Millionen Parameter, während Modelle wie Llama 7B 7 Milliarden Parameter enthält!

## Anpassungstechniken

### Prompt Engineering

Prompt Engineering bezeichnet die Art und Weise wie man Prompts schreibt. Das Ziel dabei ist einen Prompt so zu formulieren, dass das Modell effektiv damit arbeiten kann und die gewünschte Antwort generieren kann.


### Retrieval Augmented Generation (RAG)

RAG wird genutzt, um dem Modell auf gestellte Fragen noch entsprechenden Kontext geben zu können, um bessere Antworten zu bekommen.
<br>Dafür werden Informationen in einer Datenbank im Hintergrund abgespeichert. Diese können bei einer Frage an das Modell abgerufen werden und als Kontext zusätzlich zur eingegebenen Prompt übergeben werden.
<br>Dadurch kann das Sprachmodell sozusagen zur gestellten Frage in der Datenbank "nachschlagen" und somit auch eine genauere und korrektere Antwort liefern.

## Technischer Teil zum Aufbau der Chatumgebung 

<div style="display: flex; align-items: stretch; width: fit-content; border-radius: 15px; overflow: hidden; border: 2px solid #245D97; background: rgba(48, 71, 94, .1); margin-left: 20px; max-width: 800px;">

    </div>
    <div style="padding: 14px; font-family: Arial, sans-serif; font-size: 16px; line-height: 1.5; font-weight: bold; flex-grow: 1;">
        Dieser Teil ist zum Verständis nicht so wichtig. Bitte unteren Code einfach ausführen und weitergehen.
    </div>

```python
# Ollama Bibliothek wird importiert
import ollama

# Eine leere Liste als "Gedächtnis" des Modells
chat_messages = []

# Funktion, die Rolle und Nachricht bestimmt
def create_message(message, role):
  return {
    'role': role,
    'content': message
  }

# Chat-Funktion
def chat():
  # Ollama API wird aufgerufen, um Antwort vom KI-Tutor zu generieren
  ollama_response = ollama.chat(model=tutor, stream=True, messages=chat_messages)

  # Alle Antworten stückweise mit Stream ausgeben
  assistant_message = ''
  for chunk in ollama_response:
    assistant_message += chunk['message']['content']
    print(chunk['message']['content'], end='', flush=True)
    
  # Endgültige Nachricht in das Gedächtnis speichern
  chat_messages.append(create_message(assistant_message, 'assistant'))

# Frage-Funktion
def ask(message):
  chat_messages.append(
    create_message(system_prompt, 'assistant')
  )
  chat_messages.append(
    create_message(message, 'user')
  )

  print(f'\n\n--Prompt: {message}--\n\n')
  chat()

# source: https://developers-blog.org/ollama-python-library-tutorial-with-examples/
```

Nun kannst du etwas mit dem Modell rumprobieren, indem du fragen stellst wie <strong>"Erzähl mir eine Geschichte!"</strong> oder <strong>"Kennst du Bowser?"</strong>.</p>

```python
import ollama

def chat_with_ollama():
    # Stelle sicher, dass der Ollama-Server läuft
    while True:
        user_input = input("Stelle eine Frage (oder schreibe 'exit' zum Beenden): ")
        if user_input.lower() == "exit":
            print("Chat beendet.")
            break
        
        # Initiieren des Chat mit Ollama
        stream = ollama.chat(
            model='mario',
            messages=[{'role': 'user', 'content': user_input}],
            stream=True,
        )

        # Streaming der Antworten
        for chunk in stream:
            print(chunk['message']['content'], end='', flush=True)

# Starte den Chat mit Ollama
chat_with_ollama()
```

    <p>Temperature beeinflusst die Kreativität und Vorhersagbarkeit der Antworten eines Sprachmodells. Ein höherer Wert erhöht die Vielfalt und Kreativität der generierten Texte, indem ungewöhnlichere Wortkombinationen und Ideen gefördert werden. Ein niedrigerer Wert führt zu vorhersagbareren und sichereren Antworten. Dieser Parameter ist besonders nützlich, wenn man die Neigung des Modells zur Risikobereitschaft in seinen Antworten steuern möchte.</p>
    <p>Die Temperature kann festgelegt werden, indem man den Wert in der Modelfile als Parameter angibt. Dieser Wert muss sich zwischen 0 und 1 befinden.</p>
    <p>Ein Beispiel hierfür wäre:</p>
    <pre style="padding: 10px; border-radius: 5px; font-style: italic; font-weight: bold; margin: 10px 0; width: fit-content; white-space: nowrap;">
        PARAMETER temperature 0.7
    </pre>
</div>

<div style="background: rgba(255, 165, 0, .1); border-radius: 20px; padding: 20px; margin: 20px; width: fit-content; box-shadow: 0 4px 6px rgba(0,0,0,0.1); border: 2px solid orange; overflow: hidden; margin-right: 200px;">
    <p style="font-size: 22px; color: #FFF; background-color: #ffa500; padding: 10px 20px 10px 20px; margin: -20px; margin-bottom: 10px; font-weight: bold; font-style: italic;">Repeat Penalty</p>
    <p>Repeat Penalty dient dazu, Wiederholungen in der Textausgabe des Modells zu reduzieren. Ein höherer Wert für diesen Parameter bestraft Wiederholungen stärker, was zu einer erhöhten Vielfalt im generierten Text führt. Ein niedrigerer Wert macht das Modell toleranter gegenüber Wiederholungen, was nützlich sein kann, wenn Konsistenz in der Sprache oder Thematik gewünscht ist.</p>
    <p>Die Repeat Penalty kann festgelegt werden, indem man den Wert in der Modelfile als Parameter angibt. Ein hoher Wert wäre zum Beispiel 1.5, ein niedriger 0.9 und der Standardwert liegt bei 1.1.</p>
    <p>Ein Beispiel hierfür wäre:</p>
    <pre style="padding: 10px; border-radius: 5px; font-style: italic; font-weight: bold; margin: 10px 0; width: fit-content; white-space: nowrap;">
        PARAMETER repeat_penalty 1.5
    </pre>
</div>

<div style="background: rgba(255, 165, 0, .1); border-radius: 20px; padding: 20px; margin: 20px; width: fit-content; box-shadow: 0 4px 6px rgba(0,0,0,0.1); border: 2px solid orange; overflow: hidden; margin-right: 200px;">
    <p style="font-size: 22px; color: #FFF; background-color: #ffa500; padding: 10px 20px 10px 20px; margin: -20px; margin-bottom: 10px; font-weight: bold; font-style: italic;">Num Predict</p>
    <p>Num Predict legt die maximale Anzahl an Token fest, die das Modell bei der Generierung eines Textes vorhersagen darf. Dies kann verwendet werden, um die Länge der Ausgabe zu kontrollieren. Ein Wert von -1 ermöglicht eine unendliche Generierung, bis die Kontextgrenzen oder andere Limitierungen erreicht sind, während eine spezifische Zahl die Ausgabe auf eine bestimmte Anzahl von Token beschränkt.</p>
    <p>Ein Wort besteht im Schnitt aus 1,3 Tokens. Der Standardwert ist 128, bei -1 sind es unendlich viele Tokens und bei -2 wird dem Modell das maximale Kontextfenster freigegeben. Jedes Modell hat eine festgelegte maximale Kontextgröße (z.B. 2048 Tokens), dies ist aber je nach Modell unterschiedlich.</p>
    <p>Ein Beispiel hierfür wäre:</p>
    <pre style="padding: 10px; border-radius: 5px; font-style: italic; font-weight: bold; margin: 10px 0; width: fit-content; white-space: nowrap;">
        PARAMETER num_predict 1000
    </pre>
</div>

<p style="margin: 20px; color:orange; margin-bottom: 0px; margin-top: 40px;"><strong>So könnte eine Modelfile mit den entsprechenden Parametern aussehen:</strong></p>

```python
import ollama

modelfile = '''
FROM {0}
SYSTEM Du darfst nur auf Deutsch antworten! Du bist der schreckliche Bowser aus Super Mario Bros.

PARAMETER temperature 1
PARAMETER repeat_penalty 1.5
PARAMETER num_predict 1000

'''.format(modell) # In 'SYSTEM' kann der Systemprompt spezifiziert werden
                   # In 'PARAMETER' lassen sich verschiedene Parameter definieren und anpassen
                   
ollama.create(model='Bowser', modelfile=modelfile) # Erstellt das angepasste Modell 'Bowser' mit den definierten Einstellungen im 'modelfile'
```

```python
import ollama

def chat_with_ollama():
    # Stelle sicher, dass der Ollama-Server läuft
    while True:
        user_input = input("Stelle eine Frage (oder schreibe 'exit' zum Beenden): ")
        if user_input.lower() == "exit":
            print("Chat beendet.")
            break
        
        # Initiieren des Chat mit Ollama
        stream = ollama.chat(
            model='Bowser',
            messages=[{'role': 'user', 'content': user_input}],
            stream=True,
        )

        # Streaming der Antworten
        for chunk in stream:
            print(chunk['message']['content'], end='', flush=True)

# Starte den Chat mit Ollama
chat_with_ollama()
```

<h1 style="color: #ff0350; font-size: 40px; font-weight: bold;">Sprach-KI zum Fachexperten machen mit "Retrieval-Augmented Generation"</h1>

Auch mit einem optimalen Prompt kann es trotzdem noch häufig vorkommen, dass das Sprachmodell falsche oder ungenaue Antworten zurückgibt, weil es keine oder nicht genug Informationen zu dem betreffenden Thema hat.
<br>Bei diesem Problem kann RAG helfen.

<div style="width: 500px; padding: 0; border-radius: 10px; border: 3px solid #2196F3; overflow: hidden; box-shadow: 0px 4px 8px rgba(0, 0, 0, 0.2); background: rgba(33, 150, 243, .1);">
        <div style="background-color: #2196F3; color: white; height: 50px; width: 500px;">
            <p style="font-size: 22px; font-weight: bold; line-height: 45px; text-align: center;">Was ist überhaupt "RAG"?</p>
        </div>
        <ul style="list-style: none; margin: 0; padding: 14px; font-size: 16px;">
            <b>RAG</b> steht für <b>Retrieval Augmented Generation</b>, also <b>"Abruferweiterte Generierung"</b>. 
            <br><br>Damit ist gemeint, dass wir bestimmte Informationen zu dem gefragten Thema im Hintergrund z.B. von einer Datenbank (in diesem Fall eine Vektordatenbank namens "ChromaDB") abrufen und diese dann unserem Modell bei unserer Frage als Kontext mitgeben können.
            <br><br>Auf diese Weise bekommt man vom Modell viel genauere Antworten  und kann sogenannte "Halluzinationen" vermindern.
        </ul>
    </div>

Also lass uns mal versuchen, unser Modell ein wenig genauer zu machen:


Zuerst müssen wir die Bibliotheken importieren, damit wir alles haben, womit wir arbeiten wollen:

```python
from langchain_community.document_loaders.web_base import WebBaseLoader
from langchain_community.vectorstores.chroma import Chroma
from langchain_community import embeddings
from langchain_community.chat_models.ollama import ChatOllama
from langchain_core.runnables import RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import ChatPromptTemplate
from langchain.text_splitter import CharacterTextSplitter
```

Danach legen wir fest, welches Modell wir verwenden wollen. Das Modell solltest du in Ollama installiert haben.
<br>In diesem Fall verwenden wir Llama3 (wenn du ein anderes Modell nutzen willst, kannst du es in der Zelle einfügen. Beachte dabei, dass die Beispiele bei größeren Modellen wie Mixtral 8x7b nicht so anschaulich sind):

<span style="color:#2196F3">**model**</span> beschreibt, welches Sprachmodell wir verwenden wollen.
<br><span style="color:#2196F3">**num_predict**</span> steuert, wie viele Tokens das Modell maximal generieren darf, bevor die Ausgabe abgebrochen wird.
<br><span style="color:#2196F3">**temperature**</span> beschreibt, wie variabel die Ausgabe wird (Bei einem Wert näher null, ist es wahrscheinlicher, fast dieselbe Antwort bei gleicher Eingabe zu bekommen und näher an 1 werden die Antworten bei gleicher Eingabe unterschiedlicher).
<br><span style="color:#2196F3">**"stream=True"**</span> bewirkt, dass wir nicht erst warten müssen, bis die Antwort vollständig fertig generiert ist, sondern jedes generierte Token direkt sehen können (so wie du es vielleicht auch von ChatGPT kennst).
<br><span style="color:#2196F3">**stop**</span> enthält sogenannte "Stop-Tokens", die das Ende der Ausgabe markieren. Obwohl diese in der Modelfile bereits geregelt werden, müssen sie hier noch einmal explizit genannt werden, da sich das Modell sonst in einer Art Endlosschleife verfängt.

```python
model_local = ChatOllama(
    model="llama3",
    num_predict=1000,
    temperature=0.3,
    stream=True,
    stop=[
        "<|start_header_id|>",
        "<|end_header_id|>",
        "<|eot_id|>",
        "<|reserved_special_token",
    ],
)
```

<br><br>Der Text von diesen Seiten wird in sogenannte **"Chunks"** aufgeteilt, die sich mit einem **"Overlap"** überlappen. Damit "schneiden" wir uns die Webseiten sozusagen ihn handhabbare Häppchen, mit denen das Modell viel einfacher arbeiten kann.

```python
links = [
    "https://www.nachhilfe-team.net/lernen-leicht-gemacht/dreisatz-berechnen/",
    "https://www.matheretter.de/wiki/dreisatz",
    "https://www.mathebibel.de/dreisatz",
]
docs = [WebBaseLoader(link).load() for link in links]
docs_list = [item for sublist in docs for item in sublist]
text_splitter = CharacterTextSplitter(chunk_size=2000, chunk_overlap=50)
doc_splits = text_splitter.split_documents(docs_list)
```

<div style="display: flex; align-items: stretch; width: fit-content; margin: 20px; border-radius: 15px; overflow: hidden; border: 2px solid #2196F3; background: rgba(33, 150, 243, .1);">

        Lass dich auch hier wieder nicht abschrecken. Die folgende Zelle kann einfach ausgeführt werden, wenn du das Modell "nomic-embed-text" noch nicht in Ollama installiert hast. 
        <br>(Das kannst du mit der Zelle danach überprüfen)
    </div>
</div>

```python
from tqdm import tqdm
from ollama import pull


current_digest, bars = "", {}
for progress in pull("nomic-embed-text", stream=True):
    digest = progress.get("digest", "")
    if digest != current_digest and current_digest in bars:
        bars[current_digest].close()

    if not digest:
        print(progress.get("status"))
        continue

    if digest not in bars and (total := progress.get("total")):
        bars[digest] = tqdm(
            total=total,
            desc=f"pulling {digest[7:19]}",
            unit="B",
            unit_scale=True,
        )

    if completed := progress.get("completed"):
        bars[digest].update(completed - bars[digest].n)

    current_digest = digest

print("Du kannst nun mit dem Tutorial fortfahren!")
```

```python
retriever = db.as_retriever()
```

Damit wir einen guten Vergleich haben, testen wir als erstes, welche Antwort unser Modell ohne die Informationen von den Webseiten liefert.
<br><br>Dafür stellen wir unserem Modell einfach die Frage **"Wie wende ich den Dreisatz an?"** mit der vorherigen Information, dass die Frage das Schulfach Mathematik betrifft:

```python
before_rag_template = "Beantworte die folgende Frage aus dem Schulfach Mathematik anhand eines Beispiels: {question}"
before_rag_prompt = ChatPromptTemplate.from_template(
    template=before_rag_template
)
before_rag_chain = before_rag_prompt | model_local | StrOutputParser()

for chunk in before_rag_chain.stream("Wie wende ich den Dreisatz an?"):
    print(chunk, end="", flush=True)
```

Da der Dreisatz unter diesem Namen etwas sehr spezifisches in Deutschland ist und das Modell vor allem mit englischen Daten trainiert wurde, hat die Antwort, die du bekommen hast, vermutlich nicht viel mit dem tatsächlichen Dreisatz zu tun, den wir in der Schule gelernt haben.
<br><br>Wenn das Modell etwas nicht weiß, versucht es, sich den Zusammenhang zusammenzureimen und erfindet oft irgendetwas, das sehr logisch klingt aber wenn man sich auskennt, gar keinen Sinn ergibt oder etwas vollkommen anderes beschreibt. Das Modell **"halluziniert"**.
<br><br>Da die Antworten trotzdem oft gut formuliert sind, <span style="text-decoration:underline; color:orange">sollte man einem LLM nie zu 100% vertrauen und Antworten im Zweifelsfall immer gegenprüfen.</span> Es ist nie garantiert, dass alle Antworten, die dir ein Sprachmodell gibt, richtig sind.
<br><br>Um die Antworten unseres Modells nun trotzdem zu verbessern, wollen wir nochmal die selbe Frage stellen, aber unserem Modell diesmal den Kontext aus der Datenbanken mitgeben, die wir vorhin gefüllt haben.
<br><br>Schauen wir uns zuerst an, was dem Modell als Kontext mitgegeben wird, um die Frage zu beantworten.
<br><br>Dazu führen wir unseren **Retriever** mit der gewünschten Frage aus und lassen uns ausgeben, was der zurückgegebene Inhalt ist:

```python
ret_output = retriever.invoke("Wie wende ich den Dreisatz an?")

form_docs = ""

for doc in ret_output:
    form_docs += "\n\n-----------------------\n\n" + doc.page_content

print("anzahl der docs: ", len(ret_output), "\n\n")

print(form_docs)
```

Und jetzt lassen wir unser Modell mit diesem Kontext eine hoffentlich bessere Antwort generieren.

Versuch mal, die Stelle im Kontext zu finden, die das Modell vermutlich verwendet hat.

         Nicht wundern: da das Modell hier zusammen mit dem Kontext einen wesentlich größeren Prompt verarbeiten muss, dauert es in der folgenden Codezelle deutlich länger, bis eine Ausgabe generiert wird.
    </div>

```python
after_rag_template = """Beantworte die Frage aus dem Schulfach Mathematik nur mithilfe des folgenden Kontextes anhand eines Beispiels:
{context}
Frage: {question}
"""

after_rag_prompt = ChatPromptTemplate.from_template(
    template=after_rag_template
)
after_rag_chain = (
    {"context": retriever, "question": RunnablePassthrough()}
    | after_rag_prompt
    | model_local
    | StrOutputParser()
)

for chunk in after_rag_chain.stream("Wie wende ich den Dreisatz an?"):
    print(chunk, end="", flush=True)
```

## RAG

Hier hast du alles aus den Codezellen im RAG-Kapitel nochmal zusammengefasst in einer Zelle. Verändere gern die verschiedenen Parameter und schau, wie sich das Ergebnis verändert.

```python
# ===============================
# Import der Bibliotheken
# ===============================

from langchain_community.document_loaders.web_base import WebBaseLoader
from langchain_community.vectorstores.chroma import Chroma
from langchain_community import embeddings
from langchain_community.chat_models.ollama import ChatOllama
from langchain_core.runnables import RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import ChatPromptTemplate
from langchain.text_splitter import CharacterTextSplitter

# ===============================
# Das Schulfach und die Frage festlegen
# ===============================

# Lege hier fest, aus welchem Fach deine Frage ist, welche Frage du stellen möchtest
# und welche Links dein Modell als "Gedankenstütze" nutzen soll
subject = "Chemie"
question = "Was sind endotherme und exotherme Reaktionen?"
links = [
    "https://www.studysmarter.de/schule/chemie/physikalische-chemie/exotherme-reaktion/",
    "https://www.leifichemie.de/einfuehrung-die-chemie/einfuehrung-chemische-reaktion/grundwissen/energieumsatz-bei-chemischen-reaktionen",
    "https://www.lernort-mint.de/chemie/anorganische-chemie/chemische-reaktionen/exotherm-und-endotherm/",
    "https://www.chemie-schule.de/KnowHow/Exotherm",
]

print("===============================")
print("Betreffendes Schulfach: ", subject)
print("Frage: ", question)
print("Webseiten: ", [link for link in links])
print("===============================")

# ===============================
# Das Modell mit den Einstellungen auswählen
# ===============================

# Hier kannst du auch andere Modelle (diese müssen in Ollama installiert sein) ausprobieren,
# aber auch die Temperatur ändern und schauen, wie sich damit der Output verändert
model_local = ChatOllama(
    model="llama3",
    num_predict=1000,
    temperature=0.5,
    stream=True,
    stop=[
        "<|start_header_id|>",
        "<|end_header_id|>",
        "<|eot_id|>",
        "<|reserved_special_token",
    ],
)

# ===============================
# Internetquellen laden und in Chunks aufteilen
# ===============================

print("===============================")
print("Lade Inhalt der Webseiten...")
print("===============================")

docs = [WebBaseLoader(link).load() for link in links]
docs_list = [item for sublist in docs for item in sublist]

print("===============================")
print("Teile Seiteninhalte in Chunks...")
print("===============================")

# Hier kannst du testen, wie sich dein Ergebnis ändert, wenn du die Größe der Chunks
# oder des Overlaps änderst
text_splitter = CharacterTextSplitter(chunk_size=2000, chunk_overlap=50)
doc_splits = text_splitter.split_documents(docs_list)

# ===============================
# Chunks in ChromaDB einbetten
# ===============================

print("===============================")
print("Bette Chunks in Datenbank ein...")
print("===============================")

# Hier kannst die Zeile mit "persist_directory" einkommentieren, wenn du eine
# dauerhafte (persistente) Datenbank haben willst.
vectorstore = Chroma.from_documents(
    documents=doc_splits,
    collection_name="test_rag_chroma",
    embedding=embeddings.OllamaEmbeddings(model="nomic-embed-text"),
    # persist_directory="./sandbox_db",  # kommentiere das ein für eine persistente Datenbank
)

# Wenn du eine persistente Datenbank bereits erstellt hast, kannst statt dem
# Block oben, den folgenden einkommentieren, um auf die bereits existierende
# Datenbank zugreifen zu können.
"""
vectorstore = Chroma(
    persist_directory="./sandbox_db",
    embedding_function=embeddings.OllamaEmbeddings(model="nomic-embed-text"),
)
"""

retriever = vectorstore.as_retriever()

# ===============================
# Testausgabe ohne RAG
# ===============================
print("===============================")
print("Generiere Testausgabe ohne RAG...")
print("===============================")

before_rag_template = (
    "Beantworte die folgende Frage aus dem Schulfach "
    + subject
    + " anhand eines Beispiels: {question}"
)
before_rag_prompt = ChatPromptTemplate.from_template(
    template=before_rag_template
)
before_rag_chain = before_rag_prompt | model_local | StrOutputParser()

for chunk in before_rag_chain.stream(question):
    print(chunk, end="", flush=True)

# ===============================
# Ausgabe des RAG-Inhalts
# ===============================

# Dieser Teil ist für die Leserlichkeit der Ausgabe auskommentiert.
# Wenn du dir die Rückgabe der Vektordatenbank doch anzeigen lassen willst,
# entferne einfach die """-Stellen vor und nach dem Codeblock.
"""
ret_output = retriever.invoke(question)

form_docs = ""

for doc in ret_output:
    form_docs += "\n\n-----------------------\n\n" + doc.page_content

print("===============================")
print(
    "Anzahl der zurückgegebenen Dokumente: ",
    len(ret_output),
    "\n\n-----------------------\n\n",
)

print(form_docs)
print("===============================")
"""

# ===============================
# Testausgabe mit RAG
# ===============================
print("\n\n===============================")
print("Generiere Testausgabe mit RAG...")
print("===============================")

after_rag_template = (
    "Beantworte die Frage aus dem Schulfach "
    + subject
    + " nur mithilfe des folgenden Kontextes anhand eines Beispiels:\n{context}\nFrage: {question}"
)

after_rag_prompt = ChatPromptTemplate.from_template(
    template=after_rag_template
)
after_rag_chain = (
    {"context": retriever, "question": RunnablePassthrough()}
    | after_rag_prompt
    | model_local
    | StrOutputParser()
)


for chunk in after_rag_chain.stream(question):
    print(chunk, end="", flush=True)
```
